```python
# Importing dependencies
import pandas as pd
pd.set_option('display.max_columns', 500)
import numpy as np
import pymongo
from math import isnan
```

```python
# Reading file one
oec = pd.read_csv('resources/oec.csv')
oec
```

```python
# Formatting file one
oec_columns = {'PlanetIdentifier':'PlanetName'}
oec.rename(columns=oec_columns, inplace=True)
oec
```

```python
# Reading file two
caltech = pd.read_csv('resources/planets_2020.03.03_19.32.24.csv', low_memory=False)
```

```python
# Cleaning and formatting file two
ct_narrow = caltech[['pl_name', 'pl_controvflag', 'pl_pnum', 'pl_orbper', 'pl_dens', 'st_dist', 'st_teff',
                     'st_mass', 'st_rad', 'pl_angsep', 'pl_rvamp', 'pl_eqt', 'pl_insol', 'pl_occdep', 'pl_ratdor',
                     'pl_ratror', 'pl_facility', 'pl_status', 'pl_mnum', 'st_glon', 'st_glat', 'st_plx', 'st_radv',
                     'st_lum', 'st_dens', 'st_age']]

ct_columns = {'pl_name':'PlanetName', 'pl_controvflag':'ControversialFlag',
              'pl_pnum':'PlanetsInSystem', 'pl_orbper':'OrbitalPeriod', 'pl_dens':'PlanetDensity', 'st_dist':'Distance',
              'st_teff':'EffectiveTemperature', 'st_mass':'StellarMass', 'st_rad':'StellarRadius', 
              'pl_angsep':'CalcAngularSeparation', 'pl_rvamp':'RadialVelocityAmplitude', 'pl_eqt':'EquilibriumTemperature',
              'pl_insol':'InsolationFlux', 'pl_occdep':'OccultationDepth', 'pl_ratdor':'StellarRadiusDistanceRatio',
              'pl_ratror':'StellarRadiusPlanetRatio', 'pl_facility':'DiscoveryFacility', 'pl_status':'Status',
              'pl_mnum':'MoonsInSystem', 'st_glon':'GalacticLongitude', 'st_glat':'GalacticLatitude', 'st_plx':'Parallax',
              'st_radv':'RadialVelocity', 'st_lum':'StellarLuminosity', 'st_dens':'StellarDensity', 'st_age':'StellarAge'}
ct_narrow.rename(columns=ct_columns, inplace=True)
ct_narrow
```

```python
# Merging dataframes and changing NaNs to null
extrasolar_merged = pd.merge(oec, ct_narrow, how='outer', on='PlanetName', sort=True)
extrasolar_merged.fillna('', inplace=True)
extrasolar_merged
```

```python
# Converting merged dataframe to a list of dictionaries for upload to MongoDB
es_dict = extrasolar_merged.to_dict('records')
es_dict
```

```python
# Creating empty list to hold clean data, then looping through each entry and only keeping the data that has values
es_clean = []
for x in es_dict:
    clean = {key:value for key, value in x.items() if value != ''}
    es_clean.append(clean)
es_clean
```

```python
# Establishing connection and creating database and collection
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)

db = client.ETL
collection = db.extrasolar
```

```python
# Pushing results to database
collection.insert_many(es_clean)
```
